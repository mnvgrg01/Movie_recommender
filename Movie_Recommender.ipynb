```python
import pandas as pd
import tkinter as tkr
from tkinter import *
```

```python
movies = pd.read_csv('movies.csv')
movies = movies.sort_values(by='title')
rating = pd.read_csv('ratings.csv')
mov_rat= pd.merge(rating,movies,on='movieId')
pivot = mov_rat.pivot_table(index='userId',columns='title',values='rating')
```

```python
mov_rat.head()
pivot.head()
```

```python
app = tkr.Tk('Movie_Recommender')
app.title('Movie Recommender')
app.geometry('600x400')
```

```python
w = Scrollbar(app)
```

```python
myList = Listbox(app,yscrollcommand=w.set,width=50)
for i in movies['title']:
    myList.insert(END,i)
myList.pack(side='left',fill=BOTH)
w.config( command = myList.yview )
w.place(in_=myList,relx=1,relheight=1)
```

```python
def mov_recc():
    var = myList.get(myList.curselection())
    simm_corr = pivot.corrwith(pivot[var])
    simm_corr = pd.DataFrame(simm_corr,columns=['similarity'])
    simm_corr['count'] = mov_rat.groupby('title')['rating'].count()
    car = simm_corr[simm_corr['count']>40].sort_values(by = 'similarity',ascending=False).head(5)
    car_list = Listbox(app,width=35)
    sim_list = [str(i) for i in (car[car.columns[0]])]
    if sim_list[0] == 'nan' :
        car_list.insert(END,'Please Select Another one')
    else:
        for i in car.index[:3].tolist():
            car_list.insert(END,i)
    msg = tkr.Variable(app)
    msg.set('Recommendations:')
    tkr.Label(app, textvariable=msg).place(x=350,y=50)
    car_list.place(x=350,y=100)
    print(var,car)
```

```python
tkr.Button(app,text='Show Recommendations',command=mov_recc).place(x=390,y=350)
```

```python
app.mainloop()
```
